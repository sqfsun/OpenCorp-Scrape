```python
import requests
import time
import random
```

```python
httpaddress= "https://opencorpdata.com/sg?page="
urllist=[]
for page in range(1001):
        urllist.append(httpaddress+str(page))
```

```python
urllist=urllist[1:]
```

```python
company=[]
for url in urllist[0:1001]:
    try:
        uen=[]
        headers = {'User-Agent': random.choice(user_agents)}
        site=requests.get(url,headers=headers)
        print(url+' '+str(site.status_code))
        html=requests.get(url).content
        soup=BeautifulSoup(html,'lxml')
        Table=soup.find('table',{'class':'table table-striped table-hover table-condensed'})
        rows = Table.find_all('tr')
        for row in rows:
            try:
                index=row.find('a',href=True)['href']
                uen=index.split('sg/',1)[1]#get uen from the url after sg/
            except:
                pass

            cols = row.find_all('td')
            cols = [ele.text.strip() for ele in cols]
            cols.insert(0,uen)#insert uen to the row
            company.append([ele for ele in cols])


        time.sleep(2)
    except:
        print(url+' failed')
        continue
```

```python
df=[]
df=pd.DataFrame(company,columns=['UEN','Name','Address','Reg_Date','Status'])[1:]
```
